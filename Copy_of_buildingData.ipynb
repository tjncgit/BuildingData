```python
import pandas as pd
import io

df = pd.read_csv(io.BytesIO(uploaded['mortage.csv']))
```

```python
df = df.rename(columns={"affordm10": "2006-2010", "affordm11": "2007-2011", "affordm12": "2008-2012", "affordm13": "2009-2013", "affordm14": "2010-2014", "affordm15": "2011-2015", "affordm16": "2012-2016", "affordm17": "2013-2017", "affordm18": "2014-2018", "affordm19": "2015-2019", "affordm20": "2016-2020"})
df = df.drop(columns = ["CSA2020"])
df
```

```python
interval_average = df.iloc[:, 2:13].mean(axis=0) 
interval_average
interval_average_list = interval_average.values.tolist()
interval_average_list
```

```python
header_list = list(df.columns)

for i in header_list:
  if i == 'CSA2010' or i == 'OBJECTID':
    header_list.pop(1)
    header_list.remove(i)

header_list
```

list of baltimore areas

```python
areas = df.iloc[:, 1]
areas_list = list(areas)
areas_list
```

the average area list shows the average percentage of people who spend 30 percent of their income on mortage each baltimore area in all time intervals

```python
area_average = df.iloc[:, 2:13].mean(axis=1)
area_average_list = list(area_average)
area_average_list
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(header_list[0:11])
y = np.array(interval_average_list)
fig = plt.figure(figsize =(10, 7))

plt.title("average of percentage in each time interval across all baltimore areas")
plt.barh(x,y)
plt.show()
```

```python
x = np.array(areas_list)
y = np.array(area_average_list)
fig = plt.figure(figsize =(10, 15))

plt.title("average of percentage in each baltimore area across all time intervals")
plt.barh(x,y)
plt.show()
```

```python
import statistics

baltimore_average = statistics.mean(interval_average_list)
baltimore_average
```
